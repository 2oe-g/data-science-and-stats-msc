```python
def parse_data(file):
    with open(file, 'r+') as f:
        results=[]
        data = f.read()
    for line in data.split('\n'):
        a=line.replace('"','').split(',')
        results.append(a)
    return results
barometer=parse_data('barometer-last-year.csv')
indoor=parse_data('indoor-temperature-last-year.csv')
outside=parse_data('outside-temperature-last-year.csv')
rainfall=parse_data('rainfall-last-year.csv')

###############################################UNUSED CODE########################################################

#filename = 'barometer-last-year.csv'
#with open(filename, 'r') as f:
 #   for line in f:
   #     row = line.replace("\n","").replace("\ufeff","").split(',') 
   #     print(row)

#filename = 'indoor-temperature-last-year.csv'
#with open(filename, 'r') as f:
 #   for line in f:
  #      row = line.replace("\n","").replace("\ufeff","").split(',') 
         #print(row)
#filename = 'rainfall-last-year.csv'
#with open(filename, 'r') as f:
 #   for line in f:
 #      row = line.replace("\n","").replace("\ufeff","").split(',') 
         #print(row)
```

```python
monthly_df = extended_df.groupby("Month").agg([np.mean, np.std])
months = ['January', 'February', 'March', 'April','May','June', 'July', 'August','September', 'October', 'November', 'December']
monthly_df.index = pd.CategoricalIndex(monthly_df.index, categories=months, ordered=True)
monthly_df = monthly_df.sort_index()
monthly_df.reset_index(inplace=True)
monthly_df
```

```python
seasonal_df = extended_df.groupby("Season").agg([np.mean, np.std])
seasons = ['Spring','Summer','Autumn','Winter']
seasonal_df.index = pd.CategoricalIndex(seasonal_df.index, categories=seasons, ordered=True)
seasonal_df = seasonal_df.sort_index()
seasonal_df.reset_index(inplace=True)
seasonal_df

#highest rainfall recorded in autumn
#highest pressure recorded in winter
#highest humidity recorded in summer
#highed outdoor temp recorded in summer
#highest indoor temp recorded in summer

#lowest pressure recorded in autumn
#lowest humidity recorded in winter
#lowest outdoor temp recorded in autumn
#lowest indoor temp recorded in winter
```
