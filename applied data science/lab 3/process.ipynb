```python
### transforming activity description input into lowercase and also accounting for different combinations of 
#### phrases to reduce erroneous outputs

def activity_code_output(activity_description):
    
    if activity_description.lower() == "eating/drinking" or activity_description.lower() == "eating / drinking" \
    or activity_description.lower() ==  "eating/ drinking" or activity_description.lower() ==  "eating /drinking":
        return "ED145"
    elif activity_description.lower() == "education/lectures" or activity_description.lower() == "education / lectures" \
    or activity_description.lower() == "education/ lectures" or activity_description.lower() == "education /lectures":
        return "EL642"
    elif activity_description.lower() == "exercise":
        return "X893"
    elif activity_description.lower() == "housework":
        return "H179"
    elif activity_description.lower() == "leisure":
        return "L418"
    elif activity_description.lower() == "reading":
        return "R523"
    elif activity_description.lower() == "sleep":
        return "S801"
    elif activity_description.lower() == "travelling":
        return "T695"
    elif activity_description.lower() == "using devices":
        return "UD415"
    elif activity_description.lower() == "paid work":
        return "PW101"
    elif activity_description.lower() == "coursework":
        return "CW982"
    elif activity_description.lower() == "other":
        return "O733"
    else:
        return "" 
    
```

```python
def uni_bath_week_no_output(activity_date):
    
    #reformatting from dd/mm/yyyy to yyyy-mm-dd is necessary as this is the format used by .isocalendar
    
    date_object = datetime.datetime.strptime(str(activity_date), '%d/%m/%Y')
    reformatted_date = date_object.fromisoformat(date_object.strftime('%Y-%m-%d')).isocalendar()[1]  
    
    if datetime.date(2022, 2, 7).isocalendar()[1] == reformatted_date:
        return 19
    elif datetime.date(2022, 2, 14).isocalendar()[1] == reformatted_date:
        return 20
    elif datetime.date(2022, 2, 21).isocalendar()[1] == reformatted_date:
        return 21
    elif datetime.date(2022, 2, 28).isocalendar()[1] == reformatted_date:
        return 22
    elif datetime.date(2022, 3, 7).isocalendar()[1] == reformatted_date:
        return 23
    elif datetime.date(2022, 3, 14).isocalendar()[1] == reformatted_date:
        return 24
    elif datetime.date(2022, 3, 21).isocalendar()[1] == reformatted_date:
        return 25
    elif datetime.date(2022, 3, 28).isocalendar()[1] == reformatted_date:
        return 26
    elif datetime.date(2022, 4, 4).isocalendar()[1] == reformatted_date:
        return 27
    elif datetime.date(2022, 4, 11).isocalendar()[1] == reformatted_date:
        return 28
    elif datetime.date(2022, 4, 18).isocalendar()[1] == reformatted_date:
        return 29
    elif datetime.date(2022, 4, 25).isocalendar()[1] == reformatted_date:
        return 30
    elif datetime.date(2022, 5, 2).isocalendar()[1] == reformatted_date:
        return 31
    elif datetime.date(2022, 5, 9).isocalendar()[1] == reformatted_date:
        return 32
    else:
        return ""
    
```

```python
def enjoyment_score_output(enjoyment_rating):
    
    if enjoyment_rating.lower() == 'very unenjoyable':
        return -2
    elif enjoyment_rating.lower() == 'somewhat unenjoyable':
        return -1
    elif enjoyment_rating.lower() == 'neither enjoyable nor unenjoyable':
        return 0
    elif enjoyment_rating.lower() == 'somewhat enjoyable':
        return 1
    elif enjoyment_rating.lower() == 'very enjoyable':
        return 2
    else:
        return ''
```

```python
def mood_score_output(current_mood):
    
    if current_mood.lower() == 'awful':
        return -2
    elif current_mood.lower() == 'bad':
        return -1
    elif current_mood.lower() == 'so-so':
        return 0
    elif current_mood.lower() == 'good':
        return 1
    elif current_mood.lower() == 'amazing':
        return 2
    else:
        return ''
```

```python
def productivity_score_output(productivity_level):
    
    if productivity_level.lower() == 'zero productivity':
        return 0
    elif productivity_level.lower() == 'little productivity':
        return 1
    elif productivity_level.lower() == 'average productivity':
        return 2
    elif productivity_level.lower() == 'good productivity':
        return 3
    elif productivity_level.lower() == 'max productivity':
        return 4
    else:
        return ''
```

```python
access_csv = list()

for activity_index in range(18, data.shape[0]):
    
    activity = list(data.iloc[activity_index].values)

    PrimaryActivityCode = activity_code_output(activity[1])
    SecondaryActivityCode = activity_code_output(activity[2])
    
    if activity[3] != 'NaN':
        DescriptiveLabel = activity[3]
    else:
        DescriptiveLabel = ''

    
    if activity[0] != 'NaN':
        
        StartDateInput = activity[0]
        StartDate = StartDateInput.strftime('%d/%m/%Y')
        UniBathWeekNo = uni_bath_week_no_output(StartDate)
        DayOfWeek = StartDateInput.strftime('%A') # ensures that if there's a situation where an activity 
                                                    # starts on one day and ends the next, it will still appear 
                                                        # as the original start date (DayOfWeek)
        
        
    StartTime = datetime.time(int(activity[4]), int(activity[5])).strftime('%H:%M')
    EndTime = datetime.time(int(activity[6]), int(activity[7])).strftime('%H:%M')
        
        
    date_StartTime = datetime.datetime.combine(StartDateInput, datetime.time(int(activity[4]), int(activity[5])))
    date_EndTime = datetime.datetime.combine(StartDateInput, datetime.time(int(activity[6]), int(activity[7])))
    
    if date_StartTime > date_EndTime:
        date_EndTime = date_EndTime+datetime.timedelta(days=1)  # ensures that activites that continue past 
                                                                # midnight (eg. sleep) are correctly recorded wrt
                                                                # their start and end dates (ie. add a day for the 
                                                                # end date)
    
    DurationMins = int((date_EndTime - date_StartTime).total_seconds() / 60)
    
    EnjoymentScore = enjoyment_score_output(activity[8])

    MoodScore = mood_score_output(activity[9])
    
    ProductivityScore = productivity_score_output(activity[10])

    values = PrimaryActivityCode,SecondaryActivityCode,DescriptiveLabel,StartDate,UniBathWeekNo,DayOfWeek,StartTime,EndTime,DurationMins,EnjoymentScore,MoodScore,ProductivityScore
    
    access_csv.append(values)
    
```

```python
csv_df = pd.DataFrame(access_csv, columns = \
                      ['PrimaryActivityCode', 'SecondaryActivityCode', 'DescriptiveLabel', 'StartDate', \
                       'UniBathWeekNo', 'DayOfWeek', 'StartTime', 'EndTime', 'DurationMins', 'EnjoymentScore', \
                       'MoodScore', 'ProductivityScore'])
csv_df.to_csv('complete_data.csv', index=None)
```
